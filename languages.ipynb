```julia
function detect_language(text, languages=collect(keys(letter_frequencies)))
    C = Dict()
    for c in text
        C[c] = get(C, c, 0) + 1
    end
    scores = zeros(length(languages))
    println(C)
    for (c,f) in C
        for (i, lang) in enumerate(languages)
            freqmap = letter_frequencies[lang]
            if freqmap isa AbstractDict
                sc =  get(freqmap, c, 0.0)
            else
                sc = freqmap(c)
            end
            scores[i] += (f/length(text) - sc/100)^2
            println(c, " ", lang, " ", sc/100, " ", f/length(text))
        end
    end
    order = sortperm(scores, rev=false)
    println(Pair.(languages[order], scores[order]))
    print(scores)
    return languages[argmin(scores)]
end
```

```julia
function bigrams(text)
    D = Dict{String, Float64}()
    s = 0
    for line in eachline(IOBuffer(text))
        line = strip(line)
        if !isempty(line)
            for i in eachindex(line)
                if nextind(line, i, 1) <= lastindex(line)
                    bi = line[i:nextind(line, i, 1)]
                    D[bi] = get(D, bi, 0) + 1
                    s += 1
                end
            end
        end
    end
    for k in keys(D)
        D[k] /= s
    end
    return D
end
function clean_text(text)
    text = replace(text, r"https?://[-_.?&~;+=/#0-9A-Za-z]{1,2076}" => " ")
    text = replace(text, r"[-_.0-9A-Za-z]{1,64}@[-_0-9A-Za-z]{1,255}[-_.0-9A-Za-z]{1,255}" => " ")
    text = replace(text, r"[^\p{L}]" => " ")
    text = replace(text, r"\s\s+" => " ")
    return strip(lowercase(text))
end
function ngrams(text::AbstractString, n, counter=Dict{NTuple{n, UInt8}, Float64}())
	text = transcode(UInt8, string(clean_text(text)))
	for i in 1:length(text)-n+1
        p = Tuple(text[i:i+n-1])
        counter[p] = get(counter, p, 0) + 1
	end
    for k in keys(counter)
        counter[k] /= sum(values(counter))
    end
    counter
end
```

```julia
function similarity(d1, d2)
    sc = 0
    for k in keys(d1)
        if k in keys(d2)
            sc += d1[k] * d2[k]
        end
    end
    sc
end
function detect(text, langs=collect(keys(Bigrams)))
    bi0 = ngrams(text,3)
    scores = [similarity(bi0, Bigrams[lang]) for lang in langs]
    order = sortperm(scores, rev=true)
    Pair.(langs[order], scores[order])
end
```
