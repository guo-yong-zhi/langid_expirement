```julia
using Random
Random.seed!(42)
TD = TatoebaDataset("corpus/tatoeba")
ll = TD.lang_lines
for (k, v) in ll
    shuffle!(v)
end
ratio = 0.8 
train = Dict()
test = Dict()
for (k, v) in ll
    n = length(v)
    p = floor(Int, n*ratio)
    train[k] = sort(v[1:p])
    test[k] = sort(v[p+1:end])
end
```

```julia
dump_index(train, "tatoeba_train.txt")
dump_index(test, "tatoeba_test.txt")
```
