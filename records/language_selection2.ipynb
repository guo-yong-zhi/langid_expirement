```julia
ss = "./sentences.csv"
dir = "tatoeba"
open(ss) do f
    files = Dict()
    for line in eachline(f)
        if ! isempty(line)
            line = split(line, "\t")
            @assert length(line) == 3
            id, lang, text = line
            fn = "$dir/$lang.txt"
            if ! haskey(files, fn)
                files[fn] = open(fn, "w")
            end
            file = files[fn]
            write(file, text)
            write(file, "\n")
        end
    end
    flush.(values(files))
end
```

```julia
lc = [split(line, "\t")[2] for line in eachline(open(ss))]
```

```julia
print(unique(lc))
```

```julia
C = Dict()
for la in lc
    C[la] = get!(C, la, 0) + 1
end
```

```julia
sl = sort(collect(C), by = x -> x[2], rev = true)
println.(sl)
```
